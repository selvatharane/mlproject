Algorithm - XGBoost Regressor
predicting the house price in boston

XGBoost is an ensemble model that is incorporating more than one model.it is a type of decision tree model

Attribute Information
Input features in order:
1) CRIM: per capita crime rate by town
2) ZN: proportion of residential land zoned for lots over 25,000 sq.ft.
3) INDUS: proportion of non-retail business acres per town
4) CHAS: Charles River dummy variable (1 if tract bounds river; 0 otherwise)
5) NOX: nitric oxides concentration (parts per 10 million) [parts/10M]
6) RM: average number of rooms per dwelling
7) AGE: proportion of owner-occupied units built prior to 1940
8) DIS: weighted distances to five Boston employment centres
9) RAD: index of accessibility to radial highways
10) TAX: full-value property-tax rate per $10,000 [$/10k]
11) PTRATIO: pupil-teacher ratio by town
12) B: The result of the equation B=1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
13) LSTAT: % lower status of the population

Output variable:
1) MEDV: Median value of owner-occupied homes in $1000's [k$]



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset = pd.read_csv("boston.csv")
```

```python
house_price_dataset.head()
```

```python
house_price_dataset.shape
```

```python
# check for missing values

house_price_dataset.isnull().sum()
```

```python
house_price_dataset.describe()
```

```python
# understanding correlation between various features
#1.positive correlation if one feature increases the other also increases
#2.negative correlation if one feature decreases the other also decreases

correlation = house_price_dataset.corr()

plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':8}, cmap = 'Blues')

#cbar = color bar, sruare represents the values are inside square, fmt represents the float value after decimal point, annotations represents feature names,size represents the text size and finally color
```

```python
X = house_price_dataset.drop(columns = ['MEDV'],axis =1)
Y=house_price_dataset['MEDV']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error

score_1 = metrics.r2_score(Y_train,training_data_prediction)

#mean absolute error

score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)

print("R squared error:",score_1) #for good performance the value should be nearer to 0,1 if it is above or around 10 it is not performing well
print("Mean Absolute error: ",score_2)
```

Visualising the actual and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual price vs Predicted price")
plt.show()
```

```python
# error value for testing data

test_data_prediction = model.predict(X_test)
```

```python
 # R squared error

score_1 = metrics.r2_score(Y_test,test_data_prediction)

#mean absolute error

score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error:",score_1) #for good performance the value should be nearer to 0,1 if it is above or around 10 it is not performing well
print("Mean Absolute error: ",score_2)
```
